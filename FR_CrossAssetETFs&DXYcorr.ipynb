# Cross-Asset ETFs And The US Dollar Index

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use('seaborn-dark-palette')

import warnings 
warnings.filterwarnings('ignore')

import eikon as tr
tr.set_app_key('DEFAULT_CODE_BOOK_APP_KEY')
df, e= tr.get_data(['SPY','QQQ.O','DIA','EFA','EEM','TLT.O','LQD','GDX','SLV','.DXY'],['tr.priceclose.date','tr.priceclose'],{'SDate':'2010-01-01','EDate':0,'Frq':'D'})

def cleaning(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True) 
    df.index = df.index.map(lambda t: t.strftime('%Y-%m-%d'))
    df.columns = map(str.lower, df.columns)
    df.rename(columns = {'price close':'priceclose'},inplace=True)
    return df
df = cleaning(df)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15,2))
sns.heatmap(df.corr()[:1],vmin=-1,vmax=1,annot=True,linewidths=1,cmap="YlGnBu")
plt.xticks(rotation=35)
ax.set_ylabel('')    
ax.set_xlabel('')
ax.set_title('Correlation Analysis', size=15);
```
